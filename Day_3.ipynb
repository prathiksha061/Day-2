```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = "https://raw.githubusercontent.com/nunnarilabs/ml/master/house/kc_house_data.csv"
df = pd.read_csv(data)

print(df.head())
```

```python
print(df.shape)
print(df.columns)
print(df.dtypes)
df.info()
df.describe()
```

```python
print(df.isnull().sum())
```

```python
print(df.duplicated().sum())
```

```python
df = df.drop(columns=['id', 'date'])
```

```python
corr = df.corr()
print(corr)
```

```python
print(corr['price'].sort_values(ascending=False))
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(corr, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.scatterplot(x='sqft_living', y='price', data=df)
plt.title("sqft_living vs price")
plt.show()
```

```python
sns.scatterplot(x='grade', y='price', data=df)
plt.title("grade vs price")
plt.show()
```

```python
sns.scatterplot(x='bathrooms', y='price', data=df)
plt.title("bathrooms vs price")
plt.show()
```

```python
plt.hist(df['price'], bins=20)
plt.title("price distribution")
plt.show()
```

```python
sns.histplot(df['sqft_living'], kde=True)
plt.title("sqft_living Distribution")
plt.show()
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv")

print(df.columns)
```

```python
print(df.isnull().sum())
```

```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])

print(df.isnull().sum())
```

```python
df = df.drop(columns=['Cabin', 'Ticket', 'PassengerId'])
print(df.columns)
```

```python
df = df.rename(columns={
    'Pclass': 'Passenger_Class',
    'SibSp': 'Siblings_Spouses',
    'Parch': 'Parents_Children'})
print(df.columns)
```

```python
df['Age'] = df['Age'].astype(int)
df['Survived'] = df['Survived'].astype(bool)
df['Passenger_Class'] = df['Passenger_Class'].astype('category')
print(df.dtypes)
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/nunnarilabs/ml/master/automobileEDA.csv")
print(df['make'].value_counts())
```

```python
print(df['body-style'].value_counts())
```

```python
print(df['drive-wheels'].value_counts(normalize=True))
```

```python
print(df['fuel-system'].value_counts())
```

```python
avg_price_make = df.groupby('make')['price'].mean()
print(avg_price_make)
```

```python
print(avg_price_make.idxmax())
```

```python
avg_hp = df.groupby('body-style')['horsepower'].mean()
print(avg_hp)
print(avg_hp.idxmax())
```

```python
avg_mpg = df.groupby('drive-wheels')['city-mpg'].mean()
print(avg_mpg)
print(avg_mpg.idxmax())
```

```python
print(df.groupby('body-style')['price'].agg(['min','max','mean']))
```

```python
combo = df.groupby(['make','body-style'])['price'].mean()
print(combo)
print(combo.idxmax())
```

```python
corr = df.corr(numeric_only=True)
print(corr)
```

```python
price_corr = corr['price'].sort_values(ascending=False)
print(price_corr.head(4))
```

```python
print(df['horsepower'].corr(df['price']))
```

```python
print(df['city-mpg'].corr(df['price']))
```
